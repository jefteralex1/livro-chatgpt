```python
import pandas as pd
import re

def extrair_nomes(nome):
    nome_sem_numeros = re.sub(r'\d+', '', nome)
    return nome_sem_numeros.strip()

# Carregar o arquivo CSV
dados = pd.read_csv('/content/drive/MyDrive/big-data/tabeladados.csv')

# Aplicar a função à coluna "Nome"
dados['Nome'] = dados['Nome'].apply(extrair_nomes)

# Converter as colunas de datas para o tipo datetime
dados['Data de Admissão'] = pd.to_datetime(dados['Data de Admissão'])
dados['Data de Alta'] = pd.to_datetime(dados['Data de Alta'])

# Calcular a diferença de dias entre a data de alta e a data de admissão
dados['Dias de Internamento'] = (dados['Data de Alta'] - dados['Data de Admissão']).dt.days

# Preencher valores nulos na coluna "Sexo" com "M"
dados.loc[dados.Sexo.isnull(), 'Sexo'] = 'M'

# Visualizar informações sobre os dados
print(dados.shape) # linhas X colunas
print(dados.info()) # informações da base de dados
print(dados.describe()) # resumo das estatísticas das colunas que têm dados no formato int64
print(dados.isnull().sum()) # verifica valores nulos

print("\nValores na coluna 'Sexo' após preencher:")
print(dados[dados.Sexo.isnull()])

dados.head()
```
